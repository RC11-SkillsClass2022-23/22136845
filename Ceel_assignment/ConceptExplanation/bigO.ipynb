```python
import matplotlib.pyplot as plt
import numpy as np
import math

maxX = 30
maxY = 100

n = np.linspace(0, maxX)
n1 = np.linspace(0.1, maxX)

f1 = n
f2 = np.log(np.log(n))
f3 = 2**n
f4 = n**2
f5 = n * np.log(n)
f6 = np.sqrt(n)
f7 = math.e**n

fig, ax = plt.subplots(figsize=(10, 8))
ax.plot(n, f1, label = "n")
ax.plot(n, f2, label = "ln(ln(n))")
ax.plot(n, f3, label = "2**n")
ax.plot(n, f4, label = "n**2")
ax.plot(n, f5, label = "n*ln(n)")
ax.plot(n, f6, label = "sqrt(n)")
ax.plot(n, f7, label = "e**n")


ax.set_ylim([-1, maxY])

ax.grid()

# Move legend to the upper right
plt.legend(loc='upper right')

plt.show()
```

```python
import math
def Factorial(n): 
    if n == 0 : 
        return 1 
    else : 
        return int(n) * math.factorial(int(n) - 1)
    
a = input("Enter a number: ")
Factorial(a)
```
